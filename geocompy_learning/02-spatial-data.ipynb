# Geographic data in Python {#sec-spatial-class}

## 介绍

本章介绍了用于处理两种主要类型空间数据的关键 Python 包和数据结构，即：

* **shapely** 和 **geopandas** --- 用于处理矢量图层
* **rasterio** 和 **xarray** --- 用于处理栅格

正如我们将在本章后面的示例代码中看到的那样，**shapely** 和**geopandas** 是相关的：

* **shapely** 是一个“低级”包，用于处理单个矢量几何对象
* **geopandas** 是一个“高级”包，用于处理几何列（`GeoSeries` 对象），它内部包含 **shapely** 几何图形和矢量图层（`GeoDataFrame` 对象）

**geopandas** 生态系统提供了一种在 Python 中处理矢量图层的综合方法，许多包都基于它构建。
然而，栅格数据的情况并非如此：有几个部分重叠的包用于处理栅格数据，每个包都有自己的优点和缺点。
在本书中，我们关注最突出的一个：

* **rasterio** --- 一个面向空间的包，专注于“简单”栅格格式（例如 GeoTIFF），使用 `numpy` 数组和元数据对象（`dict`）的组合表示栅格 指定数组的空间参考

另一个值得一提的与栅格相关的包是 **xarray**。 它是用于处理带标签数组的通用包，因此有利于处理“复杂”栅格格式（例如 NetCDF），使用其自己的本机类表示栅格，即 `xarray.Dataset` 和 `xarray.DataArray`

本章将简要介绍基本的地理数据模型：矢量和栅格。
在演示它们在 Python 中的实现之前，我们将介绍每个数据模型背后的理论及其主导的学科。

矢量数据模型使用点、线和多边形表示世界。
它们具有离散的、明确定义的边界，这意味着矢量数据集通常具有很高的精度（但不一定准确）。
栅格数据模型将表面划分为大小不变的像元。
栅格数据集是网络制图中使用的背景图像的基础，并且自航空摄影和基于卫星的遥感设备诞生以来一直是地理数据的重要来源。
栅格将特定于空间的要素聚合到给定的分辨率，这意味着它们在空间上是一致的并且可扩展（许多全球栅格数据集都可用）。

使用哪个？ 答案可能取决于您的应用领域以及您有权访问的数据集：

* 矢量数据集和方法主导了社会科学，因为人类住区和过程（例如交通基础设施）往往具有离散的边界
* 由于对遥感数据的依赖，栅格数据集和方法在许多环境科学中占据主导地位

在某些领域有很多重叠，栅格和矢量数据集可以一起使用：例如，生态学家和人口统计学家通常同时使用矢量和栅格数据。
此外，还可以在两种形式之间进行转换
无论您的工作更多地使用矢量数据集还是栅格数据集，在使用它们之前了解底层数据模型都是值得的，如后续章节所述。
本书侧重于构建在 **geopandas** 和 **rasterio** 包上的方法，分别用于处理矢量数据和栅格数据集。

## Vector data

地理矢量数据模型基于位于坐标参考系统 (CRS) 内的点。
点可以表示独立的特征（例如，公共汽车站的位置），或者它们可以连接在一起形成更复杂的几何图形，例如线和多边形。
大多数点几何仅包含两个维度（3 维 CRS 包含一个额外的 $z$ 值，通常表示海拔高度）。

例如，在这个系统中，伦敦可以用坐标“(-0.1, 51.5)”表示。
这意味着它的位置在原点以东 -0.1 度和以北 51.5 度。
在这种情况下，原点位于地理（'lon/lat'）CRS 中的 0 度经度（本初子午线）和 0 度纬度（赤道）。
同一点也可以在英国国家电网中“东/北”值为“(530000, 180000)”的预计 CRS 中近似计算，这意味着伦敦位于东 530 公里和北 180 公里。 CRS的起源。
国家电网的起源地位于西南半岛以外的海域，这确保了英国的大部分地区都具有正的东向和北向值。



**geopandas** 为地理矢量数据提供类，并为 Python 中可重现的地理数据分析提供一致的命令行界面。
**geopandas** 为三个成熟的地理计算库提供了一个接口，它们结合起来代表了许多地理应用程序（包括 QGIS 和 R 的空间生态系统）构建的坚实基础：

* GDAL，用于读取、写入和操作各种地理数据格式，在第 8 章中介绍
* PROJ，一个强大的坐标系转换库，它是第 7 章内容的基础
* GEOS，一种平面几何引擎，用于计算具有投影 CRS 的数据的缓冲区和质心等操作，在第 5 章中介绍

与这些地理库的紧密集成使可重现的地理计算成为可能：使用 Python 等高级语言访问这些库的一个优点是您不需要了解低级组件的复杂性，从而可以专注于方法而不是 执行。
本节介绍 **geopandas** 类，为后续章节做准备（第 5 章和第 8 章分别介绍了 GEOS 和 GDAL 接口）。

### Vector data classes

在 Python 中处理地理矢量数据的主要类是分层的，这意味着最高级别的“矢量层”类由更简单的“几何列”和单独的“几何”组件组成。
本节按顺序介绍它们，从最高级别的类开始。
对于许多应用程序，高级矢量图层类（本质上是具有几何列的数据框）是所需要的。
但是，了解矢量地理对象的结构及其组成部分对于更高级的应用程序非常重要。
Python 中的三个主要矢量地理数据类是：

* `GeoDataFrame`，一个表示矢量图层的类，其中一个列是几何列（类 `GeoSeries`）
* `GeoSeries`，用于表示`GeoDataFrame`对象中的几何列的类
* `shapely` 几何对象，表示单个几何图形，例如点或多边形

前两个类（`GeoDataFrame` 和 `GeoSeries`）在 **geopandas** 中定义。
第三个类在 **shapely** 包中定义，它处理单个几何图形，并且是 **geopandas** 包的主要依赖项。

### Vector layer from scratch

在前面的部分中，我们从现有 Shapefile 中的矢量图层 (`GeoDataFrame`) 开始，并将其“分解”以提取几何列（`GeoSeries`、@sec-geometry-columns）和单独的几何图形（`shapely` ，请参阅@sec-geometries）。
在本节中，我们将演示相反的过程，从“shapely”几何构造一个“GeoDataFrame”，并组合成一个“GeoSeries”。
这会：

* 帮助您更好地理解 `GeoDataFrame` 的结构，以及
* 当您需要以编程方式构建简单的矢量图层时可能会派上用场，例如两个给定点之间的线等。

矢量图层由两个主要部分组成：几何图形和非地理属性。 图 ... 显示了如何创建“GeoDataFrame”对象——几何图形来自“GeoSeries”对象，而属性则来自“Series”对象。

（图如https://geocompr.robinlovelace.net/spatial-class.html#fig:02-sfdiagram）

非地理属性表示要素的名称或其他属性，例如测量值、组和其他事物。 为了说明属性，我们将表示 2017 年 6 月 21 日伦敦的温度为 25°C。此示例包含一个几何图形（坐标）和具有三个不同类（地名、温度和日期）的三个属性。 `GeoDataFrame` 类的对象通过将属性 (`Series`) 与简单要素几何列 (`GeoSeries`) 组合来表示此类数据。 首先，我们创建一个点几何，我们从@sec-geometries 知道如何做：

````python
lnd_point = shapely.geometry.Point(0.1, 51.5)
lnd_point
````

接下来，我们创建一个包含点的“GeoSeries”（长度为 1）。 请注意，“GeoSeries”存储 CRS 定义，在本例中为 WGS84（使用其 EPSG 代码“4326”定义）。 另请注意，“shapely”几何图形进入“列表”，以说明可以有多个（与此示例不同）：

````python
lnd_geom = gpd.GeoSeries([lnd_point], crs=4326)
lnd_geom
````

接下来，我们将 `GeoSeries` 与其他属性组合成一个 `dict`。 几何列是一个名为“geometry”的“GeoSeries”。 其他属性（如果有的话）可以使用 `list` 或 `Series` 对象来定义。 在这里，为简单起见，我们使用 `list` 选项来定义三个属性 `name`、`temperature` 和 `date`。 再次注意，`list` 的长度可以大于 1，以防我们创建的层具有多个特征：

````python
d = {
  'name': ['London'],
  'temperature': [25],
  'date': ['2017-06-21'],
  'geometry': lnd_geom
}
````

最后，可以将 `dictionary` 转换为 `DataFrame`：

````python
lnd_layer = gpd.GeoDataFrame(d)
lnd_layer
````

刚刚发生了什么？ 首先，坐标用于创建简单的特征几何体（“shapely”）。 其次，几何被转换为一个简单的要素几何列（`GeoSeries`），带有 CRS。 第三，属性与“GeoSeries”相结合。 这会生成一个名为“lnd_layer”的“GeoDataFrame”对象。

只是为了说明如何创建具有多个特征的图层，下面是一个示例，其中我们创建了一个具有两个点（伦敦和巴黎）的图层：

````python
lnd_point = shapely.geometry.Point(0.1, 51.5)
paris_point = shapely.geometry.Point(2.3, 48.9)
towns_geom = gpd.GeoSeries([lnd_point, paris_point], crs=4326)
d = {
  'name': ['London', 'Paris'],
  'temperature': [25, 27],
  'date': ['2017-06-21', '2017-06-21'],
  'geometry': towns_geom
}
towns_layer = gpd.GeoDataFrame(d)
towns_layer
````

以下表达式使用结果创建交互式地图：

## Raster data 栅格数据

### Introduction

如上所述，在 Python 中处理栅格不太围绕一个综合包进行组织（与矢量图层和 **geopandas** 的情况相比）。
相反，几个包提供了处理栅格数据的方法的替代子集。

**rasterio** 和 **xarray** 包提供了在 Python 中处理栅格的两种最著名的方法。
正如我们很快就会看到的，它们在范围和底层数据模型上有所不同。
具体来说，**rasterio** 将栅格表示为与保存空间元数据的单独对象关联的 **numpy** 数组。
然而，**xarray** 包表示具有本机“DataArray”对象的栅格，它是 **numpy** 数组的扩展，旨在将轴标签和属性与栅格数组一起保存在同一对象中 值。

这两个包都不像 **geopandas** 那样详尽无遗。
例如，当使用 **rasterio** 时，一方面，可能需要更多包来完成常见任务，例如区域统计（包 **rasterstats**）或计算地形指数（包 **richdem**） .
另一方面，通过 **rioxarray** 和 **xarray-spatial** 包扩展了 **xarray** 以容纳核心包本身缺失的空间运算符。

在接下来的两节中，我们将介绍 **rasterio**，这是我们将在本书其余部分使用的与光栅相关的包。

### 使用 **rasterio** {#sec-using-rasterio}

要使用 **rasterio** 包，我们首先需要导入它。
我们还导入 **numpy**，因为底层栅格数据存储在 **numpy** 数组中。
为了有效地处理这些问题，我们公开了所有 **numpy** 函数。
最后，我们从 rasterio.plot 子模块中导入 show 函数，以快速可视化栅格。

````python
import numpy as np
import rasterio
from rasterio.plot import show
import subprocess
````

栅格通常是从现有文件导入的。
使用 **rasterio** 时，导入栅格实际上是一个两步过程：

* 首先，我们使用 `rasterio.open` 打开一个光栅文件“连接”
* 其次，我们使用 .read 方法从连接中读取栅格值

这种分离类似于用于从文件读取的基本 Python 函数，例如用于从文本文件读取的“open”和“.readline”。
理由是我们并不总是希望将文件中的所有信息读入内存，这一点尤为重要，因为栅格大小可能大于 RAM 大小。
因此，第二步（`.read`）是有选择性的。 例如，我们可能只想读取一个栅格波段而不是读取所有波段。

在第一步中，我们将一个文件路径传递给 `rasterio.open` 函数以创建一个 `DatasetReader` 文件连接。
对于此示例，我们使用表示锡安国家公园高程的单波段栅格：

````python
src = rasterio.open('data/srtm.tif')
src
````

为了获得栅格值的第一印象，我们可以使用“show”函数绘制栅格：

````python
show(src);
````

`DatasetReader` 包含栅格元数据，即栅格值以外的所有信息。
让我们检查一下：

````python
src.meta
````

重要的是，我们可以看到：

*栅格数据类型（`dtype`）
* 栅格尺寸（`width`、`height` 和 `count`，即层数）
* 光栅坐标参考系统 (`crs`)
* 栅格仿射变换矩阵（`transform`）

最后一项（即“transform”）值得更多关注。
要在地理空间中定位栅格，除了 CRS 之外，我们还必须指定栅格 *origin* ($x_{min}$, $y_{max}$) 和分辨率 ($delta_{x}$, $delta_{y }$).
在变换矩阵表示法中，这些数据项存储如下：

```{text}
Affine(delta_x, 0.0, x_min,
       0.0, delta_y, y_max)
```

请注意，按照惯例，栅格 y 轴原点设置为最大值（$y_{max}$）而不是最小值，因此，y 轴分辨率（$delta_{y}$）为负。

最后，使用 `DatasetReader` 的 `.read` 方法读取实际的栅格值。
重要的是，我们可以阅读：

* 一个特定的层，传递一个数字索引（如在 .read(1) 中）
* 层的子集，传递索引的“列表”（如“.read（[1,2]）”）
* 所有层（如在 .read() 中）

请注意，层索引从“1”开始，这与第一个索引为“0”的 Python 约定相反。

结果对象是一个 **numpy** 数组，具有两个或三个维度：

* *三维*维度，当读取超过一层时（例如，`.read()` 或 `.read([1,2])`）。
在这种情况下，尺寸模式是“（层、行、列）”
* *两个*维度，当读取一个特定层时（例如，`.read(1)`）

让我们使用文件连接对象 src 从 srtm.tif 栅格中读取第一个（也是唯一一个）图层，例如：

````python
s = src.read(1)
s
````

结果 s 是一个二维 numpy 数组。

### Raster from scratch

在本节中，我们将演示从头开始创建栅格。 我们将创建两个小栅格，`elev` 和 `grain`，我们将在本书后面的示例中使用它们。 与创建矢量图层不同，实际上很少需要从头开始创建栅格，因为很难以编程方式将栅格与正确的空间范围对齐（GIS 软件更适合这项工作）。 尽管如此，这些示例对于更加熟悉 `rasterio` 数据结构还是很有用的。

栅格基本上是一个结合了地理配准信息的数组，即：

* 转换矩阵（将像素索引与坐标联系起来）
* CRS 定义

因此，要创建栅格，我们首先需要一个包含值的数组，然后用地理配准信息补充它。 让我们创建数组“elev”和“grain”。 `elev` 数组是一个 6 x 6 数组，具有从 1 到 36 的连续值。它可以按如下方式创建：

````python
elev = np.arange(1, 37, dtype=np.uint8).reshape(6, 6)
elev
````

`grain` 数组表示具有值 `0`、`1`、`2` 的分类栅格，分别对应于类别“粘土”、“淤泥”、“沙子”。 我们将从像素值的特定排列创建 if，如下所示：

````python
v = [1, 0, 1, 2, 2, 2, 0, 2, 0, 0, 2, 1, 0, 2, 2, 0, 0, 2, 0, 0, 1, 1,
       1, 1, 1, 1, 1, 2, 1, 1, 2, 1, 2, 2, 0, 2]
grain = np.array(v, dtype=np.uint8).reshape(6, 6)
grain
````

请注意，在这两种情况下，我们都使用“uint8”（8 位无符号整数，即“0-255”）数据类型，它至少足以表示给定栅格的所有可能值。 这是最小内存占用的推荐方法。

缺少的是光栅变换（参见@sec-using-rasterio）。 在这种情况下，由于栅格是任意的，我们还设置了一个任意变换矩阵，其中：

* 原点 ($x_{min}$, $y_{max}$) 位于 `-1.5,1.5`，并且
* 和分辨率 ($delta_{x}$, $delta_{y}$) 是 `0.5,-0.5`。

在代码方面，我们使用 `rasterio.transform.from_origin` 执行如下操作：

````python
new_transform = rasterio.transform.from_origin(west=-1.5, north=1.5, xsize=0.5, ysize=0.5)
new_transform
````

请注意，令人困惑的是，$delta_{y}$ 在 `rasterio.transform.from_origin` 中使用正值 (`0.5`) 定义，即使它最终为负值 (`-0.5`)！

现在可以在其坐标系中绘制栅格，将数组与转换矩阵一起传递给“显示”：

````python
show(elev, transform=new_transform);
````

可以用相同的方式绘制“grain”栅格，因为我们也将为其使用相同的变换矩阵：

````python
show(grain, transform=new_transform);
````

此时，我们可以使用 `rasterio` 处理栅格：

* 在真正的光栅像素坐标很重要的地方传递变换矩阵（例如在上面的函数“show”中）
* 请记住，我们在分析中使用的任何其他层都在这些坐标的相同 CRS 中

最后，要导出栅格以进行永久存储以及 CRS 定义，我们需要执行以下步骤：

* 创建一个光栅文件连接（我们在其中设置转换和 CRS，以及其他设置）
* 将具有栅格值的数组写入连接
* 关闭连接

在 `elev` 的情况下，我们按如下方式进行：

````python
new_dataset = rasterio.open(
    'output/elev.tif', 'w', 
    driver='GTiff',
    height=elev.shape[0],
    width=elev.shape[1],
    count=1,
    dtype=elev.dtype,
    crs=4326,
    transform=new_transform
)
new_dataset.write(elev, 1)
new_dataset.close()
````

请注意，我们（任意）为“elev”栅格设置的 CRS 是 WGS84，根据 EPSG 代码使用“crs=4326”定义。

下面是我们如何导出 `grain` 栅格，使用几乎完全相同的代码，只是将 `elev` 替换为 `grain`：

````python
new_dataset = rasterio.open(
    'output/grain.tif', 'w', 
    driver='GTiff',
    height=grain.shape[0],
    width=grain.shape[1],
    count=1,
    dtype=grain.dtype,
    crs=4326,
    transform=new_transform
)
new_dataset.write(grain, 1)
new_dataset.close()
````

如果栅格导出代码不清楚，请不要担心。 我们会在@read-write中详细说明栅格输出的细节。

结果，文件 `elev.tif` 和 `grain.tif` 被写入到 `output` 目录中。 这些与我们稍后在示例中使用的 data 目录中的 elev.tif 和 grain.tif 文件相同（例如，@sec-raster-subsetting）。

请注意，`elev` 和 `grain` 的变换矩阵和维度是相同的。 这意味着栅格是重叠的，并且可以组合成一个双波段栅格，结合栅格代数运算（@sec-map-algebra）等。

## 坐标参考系统

矢量和栅格空间数据类型共享空间数据固有的概念。
也许其中最基本的是坐标参考系统 (CRS)，它定义了数据的空间元素与地球（或其他物体）表面的关系。
正如本章开头介绍的那样，CRS 是地理的或预测的（请参阅...）。
本节解释每种类型，为第 7 章奠定基础，第 7 章深入探讨 CRS 的设置、转换和查询。

### 地理坐标系

地理坐标系使用两个值——经度和纬度来标识地球表面上的任何位置（参见图的左面板...）。 经度是东西方向上与本初子午线平面成角距离的位置，而纬度是赤道平面以北或以南的角距离。
因此，地理 CRS 中的距离不是以米为单位测量的。
这具有重要的后果，如第 7 节所示。

球面或椭圆面表示地理坐标系中的地球表面。
球形模型假定地球是一个给定半径的完美球体——它们具有简单的优点，但同时它们也不准确：地球不是球体！
椭球模型由两个参数定义：赤道半径和极半径。
这些是合适的，因为地球是压缩的：赤道半径比极地半径长约 11.5 公里。


椭圆体是 CRS 更广泛组成部分的一部分：基准面。
这包含有关使用什么椭球体以及笛卡尔坐标与地球表面位置之间的精确关系的信息。
有两种类型的基准面——地心基准面（例如 WGS84）和局部基准面（例如 NAD83）。
您可以在图 .... 中看到这两种类型基准的示例。
黑线代表一个地心基准，其中心位于地球的重心并且没有针对特定位置进行优化。
在显示为紫色虚线的局部基准中，椭圆表面被移动以与特定位置的表面对齐。
这些允许在本地 CRS 中考虑地球表面的局部变化，例如大山脉。


### 投影坐标参考系统

所有投影 CRS 都基于上一节中描述的地理 CRS，并依靠地图投影将地球的三维表面转换为投影 CRS 中的东向和北向（x 和 y）值。
投影 CRS 基于隐式平面上的笛卡尔坐标。

它们有一个原点、x 轴和 y 轴，以及一个线性测量单位，例如米。

如果不添加一些变形，则无法完成此过渡。
因此，地球表面的一些属性在这个过程中发生了扭曲，例如面积、方向、距离和形状。
投影坐标系只能保留其中的一个或两个属性。
投影通常根据它们保留的属性命名：等面积保留面积，方位角保留方向，等距保留距离，共形保留局部形状。

投影类型主要分为三组：圆锥、圆柱和平面（方位角）。
在圆锥投影中，地球表面沿着单条切线或两条切线投影到圆锥体上。
失真沿着切线最小化，并随着与该投影中这些线的距离的增加而增加。
因此，它最适合中纬度地区的地图。
圆柱投影将表面映射到圆柱体上。
也可以通过沿单条切线或两条切线接触地球表面来创建此投影。
在绘制整个世界时，最常使用圆柱投影。
平面投影将数据投影到一个平面上，该平面在一点或沿切线接触地球。
它通常用于绘制极地区域。

与大多数开源地理空间软件一样，**geopandas** 和 **rasterio** 软件包使用 PROJ 软件进行 CRS 定义和计算。
**pyproj** 包是 PROJ 的低级接口。
使用它的功能，我们可以检查支持的投影列表：

````python
import pyproj
epsg_codes = pyproj.get_codes('EPSG', 'CRS')  ## List of supported EPSG codes
epsg_codes[:5]  ## print first five
````

````python
pyproj.CRS.from_epsg(4326)  ## Printout of WGS84 CRS (EPSG:4326)
````

可以在“Map Projections”（1993 年）和 <https://www.geo-projections.com/> 中找到不同投影、它们的类型、属性和适用性的快速总结。

## Units

CRS 的一个基本特征是它们包含有关空间单位的信息。
显然，了解房屋的测量单位是英尺还是米是至关重要的，这同样适用于地图。
在地图上添加比例尺或其他一些距离指示器以展示页面或屏幕上的距离与地面上的距离之间的关系是一种很好的制图实践。
同样，用户了解几何坐标的单位也很重要，以确保后续计算在正确的上下文中完成。

**geopandas** 和 **rasterio** 中的 Python 空间数据结构本身不支持测量概念。
矢量图层或栅格的坐标是纯数字，指的是任意平面。
例如，根据 srtm.tif 的 .transform 矩阵，我们可以看到光栅分辨率为 0.000833，其 CRS 为 WGS84（EPSG：4326）。
我们可能知道（或可以发现）WGS84 的单位是十进制。
但是，该信息未编码在任何数字计算中。

````python
src.meta
````

因此，我们需要了解与我们合作的 CRS 单位。 通常，在地理 CRS 中这些是十进制度数，或者在预计 CRS 中是 $m$，但也有例外。
长度、面积或距离等几何计算以 CRS 的相同单位（例如 $m$ 或 $m^2$）返回纯数字。
由用户决定以何种单位给出结果并相应地处理结果。
例如，如果面积输出为 $m^2$，而我们需要 $km^2$ 的结果，则需要将结果除以 $1000^2$。
